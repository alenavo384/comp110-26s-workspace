Read a CSV of data from our data directory (folder) and format is as a list[dict[str,str]].

```python
from csv import DictReader

def read_csv_rows(filename: str) -> list[dict[str, str]]:
    result: list[dict[str, str]] = []

    # Open the csv so we can read the data!
    file_handle = open(filename, "r", encoding="utf8")

    # Read the CSV data!
    csv_reader = DictReader(file_handle)

    # Read the data from the CSV, row by row
    for row in csv_reader:
        # Append the row to our result list!
        result.append(row)

    # Close the file
    file_handle.close()
    

    # Reading from the CSV, adding the data to our results variable

    return result

read_csv_rows("data/walter_wally.csv")
```
